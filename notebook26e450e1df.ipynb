## splitting the data as target and label data

```python
x=data.drop(["Class"] , axis =1)
y=data["Class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test , y_train ,y_test = train_test_split(x , y , test_size=0.3 ) 
```

## importing all the classification models 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from xgboost import XGBClassifier
```

##  import metrics for evaluating the models

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
lr=LogisticRegression()
rfc=RandomForestClassifier()
dtc=DecisionTreeClassifier()
xg = XGBClassifier()
```

```python
lr.fit(x_train , y_train)
```

## doing prediction using logistic regression 

```python
pred= lr.predict(x_test)
```

## checking the score and accuracy of the model

```python
print(accuracy_score(y_test , pred))
```

```python
print(classification_report(y_test , pred))
```

## checking for overfitting.....
clearly the cross validation score is more similar with respect to the cv score  i.e training and test data is performing similar so it can said that the model is trained properly and it can predict well....

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr, x, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean CV score:", scores.mean())
```

##  similarly cheking for the random forest classifier 

```python
rfc.fit(x_train , y_train)
pred1=rfc.predict(x_test)
print(accuracy_score(y_test , pred1))
```

## random forest classifier  doesn't perform well as it has overfitting condition..

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(rfc, x, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean CV score:", scores.mean())
```

## predicting using decision tree

```python
dtc.fit(x_train , y_train)
pred2=dtc.predict(x_test)
print(accuracy_score(y_test , pred2))
```

## decision tree gives overtraining data

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(dtc, x, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean CV score:", scores.mean())
```

## checking with the xgboost and their performance

```python
xg.fit(x_train , y_train)
pred3=xg.predict(x_test)
print(accuracy_score(y_test , pred3))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(xg, x, y, cv=5)
print("Cross-validation scores:", scores)
print("Mean CV score:", scores.mean())
```
